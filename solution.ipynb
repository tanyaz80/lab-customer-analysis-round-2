```python
nulls_data = pd.DataFrame(round(file.isna().sum()/len(file),4)*100) # shows the percentage of null values in a column
nulls_data = nulls_data.reset_index()
display(nulls_data)

# a lot of missing data in vehicle type 
```

```python

nulls_data = nulls_data.rename(columns = {'index': 'header_name', 0:'percent_nulls'})
display(nulls_data.head())
#print(nulls_df.columns)
```

```python
#as there is a lot of missing data in vehicle type and we have better data in vehicle class and vehicle size

# I decided to remove the whole column verhicle type. After I reset index.
#Also I am removing the column "unnamed:0"
filedrop1 = file.drop(['vehicle type','unnamed: 0'], axis=1)

display(filedrop1.head())
filedrop1=filedrop1.reset_index()
filedrop1.shape # 10910,25
```

```python

#filedrop2 = filedrop1.dropna() If I drop all na elements, I got half data - shape of new data - (4543, 26).Too bad.
#As objects are harder to predict, and I could see before that vehicle class missing data at the same time as
#vehicle size, I am dropping 'vehicle class'.
#As well I decided to drop the rows with state(it's same rows that responses were missins)
#filedrop2 = filedrop1[filedrop1['vehicle class'].isna()==False] # Since these nulls are not a lot, we can filter them
rows_to_drop = filedrop1[filedrop1['vehicle class'].isna()==True].index

filedrop2 = filedrop1.drop(rows_to_drop, axis=0)
print(filedrop2.shape) #10288,25
display(filedrop2.head())
```

```python
rows_to_drop = filedrop2[filedrop2['state'].isna()==True].index
#rows_to_drop1 = filedrop1[filedrop1['state'].isna()==True].index

filedrop3 = filedrop2.drop(rows_to_drop, axis=0)
print(filedrop3.shape) #9688,25
display(filedrop3.head())
```

```python
filedrop3[filedrop3['number of open complaints'].isna()==True].head(60)
#filedrop['number of open complaints'].unique() # array([ 0., nan,  3.,  1.,  2.,  4.,  5.])
#filedrop['number of open complaints'].nunique() #6
#I the average of number of complaints is 0,38 So I will round it to 0
# I will calculate mean value for Nan values 'months since the last claim', also rounding till integer.
```

```python

mean_number_of_open_complaints = int(filedrop3['number of open complaints'].mean())
mean_number_of_open_complaints #0.38 is close to 0
filedrop3['number of open complaints'] = filedrop3['number of open complaints'].fillna(mean_number_of_open_complaints)
```

```python

filedrop3['months since last claim'].unique() 
# array(array([ 7.,  3., 34., 10., nan,  8., 35., 33., 19., 13., 24., 25.,  6.,
#       20., 26.,  5., 14.,  9., 29., 11.,  1., 18., 16., 30., 12.,  2.,
#       4., 22., 31., 21., 28., 17., 15., 32., 23., 27.,  0.]))

mean_number_months_since_last_claim = int(filedrop3['months since last claim'].mean())
mean_number_months_since_last_claim #15.097 becomes 15
filedrop3['months since last claim'] = filedrop3['months since last claim'].fillna(mean_number_months_since_last_claim)
```
